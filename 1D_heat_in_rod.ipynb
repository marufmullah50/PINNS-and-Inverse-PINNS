```python
#Library
import numpy as np
import matplotlib.pyplot as plt

# Boundary Condition
length = 10
k = 0.89
# B.C
temp_left = 100
temp_right = 200
total_sim_time = 10

# Mesh Step (SPATIAL)
dx = 0.1 #STEP size
x_vector = np.linspace(0, length, int(length/dx))

#Mesh step Time
dt = 0.0001
t_vector = np.linspace(0, total_sim_time, int(total_sim_time/dt))

print(x_vector)
print(t_vector)

# Initial Condition
u = np.zeros([len(t_vector),len(x_vector)])
u[:,0] = temp_left
u[:,-1]= temp_right

#Ploting upto this
plt.plot(x_vector,u[0])
plt.ylabel("temp")
plt.xlabel("x")
plt.show()


```

```python
#Simulation Code

for t in range(1, len(t_vector)-1):
    for x in range(1, len(x_vector)-1):
        u[t+1,x] = ((k* (dt/dx**2)) * (u[t, x+1] - 2* u[t,x] + u[t,x-1])) + u[t,x]

print(u)
```
